**Get All Books by Ian McEwan from OpenLibrary API**

```python
import requests
import json

response = requests.get('https://openlibrary.org/search/authors.json?q=Ian%20McEwan')
response
results = response.json()
results
```

```python
import pandas as pd
pd.set_option("display.max_rows", 0)

books = pd.DataFrame(results)
books
```

```python
def flatten(data, col_list):
    for column in col_list:
        flattened = pd.DataFrame(dict(data[column])).transpose()
        columns = [str(col) for col in flattened.columns]
        flattened.columns = [column + '_' + colname for colname in columns]
        data = pd.concat([data, flattened], axis=1)
        data = data.drop(column, axis=1)
    return data
```

**Get All England Bank Holidays**

```python
response = requests.get('https://www.gov.uk/bank-holidays.json')
response
results = response.json()
results
```

```python
holidays = pd.DataFrame(results).transpose()
```

```python
holidays
```

```python
from pandas.io.json import json_normalize

flattened_data = json_normalize(holidays['events'])

flattened_data
```

```python
flattened_data = flattened_data.transpose()
flattened_data = flattened_data.iloc[:76,1:2]
flattened_data
```

```python
flattened_data.columns = ['england-and-wales']
```

```python
holidays = flatten(flattened_data, ['england-and-wales'])
```

```python
holidays.drop(holidays.tail(1).index,inplace=True)
```

```python
holidays.drop(holidays.iloc[:,2:4], axis=1, inplace=True)
holidays
```
